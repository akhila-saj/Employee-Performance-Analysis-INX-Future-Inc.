```python
# Plotting EmpJObRole with respect to PerformanceRating
plt.figure(figsize=(30,20))
ax = sns.countplot(df.EmpJobRole,hue=df.PerformanceRating)
total = float(len(df))
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='right')
plt.show()
```

#### Insights:
- As we can visualize that Sales Executive role has 5.3% of low performance rating followed by Research Scientist and Manager R&D with 1.4%.
- Improvements should be made with these roles and departments to improve performance rating.

```python
# Plotting EmpEnvironmentSatisfaction variable with respect to performance rating
plt.figure(figsize=(10,10))
ax = sns.countplot(df.EmpEnvironmentSatisfaction,hue=df.PerformanceRating)
total = float(len(df))
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='right')
plt.show()
```

###  c. Correlation Between Independent Variables
Correlation expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It's a common tool for describing simple relationships without making a statement about cause and effect. So there should be least correlation between the independant variable.

```python
# Check Correlation between the variable.
plt.figure(figsize=(22,20))
cor=df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

### Insights:

- Age:-- highly correlated with EmpJobLevel, TotalWorkExperienceInYears and ExperienceYearsAtThisCompany (Can be removed)
- EmpJobLevel :-- highly correlated with Age, TotalWorkExperienceInYears and ExperienceYearsAtThisCompany
- ExperienceYearsAtThisCompany :-- EmpJobLevel, TotalWorkExperienceInYears, ExperienceYearsInCurrentRole,                                                          YearsSinceLastPromotion, YearsWithCurrManager             (REMOVE)

- ExperienceYearsInCurrentRole :-- ExperienceYearsAtThisCompany, YearsSinceLastPromotion, YearsWithCurrManager
- YearsSinceLastPromotion :-- ExperienceYearsAtThisCompany, EmpJobLevel, TotalWorkExperienceInYears,                                                         ExperienceYearsInCurrentRole,YearsWithCurrManager 	
- YearsWithCurrManager :-- EmpJobLevel, TotalWorkExperienceInYears,ExperienceYearsAtThisCompany, ExperienceYearsInCurrentRole,                             YearsSinceLastPromotion

### d. Correlation of Target and Predictor Variables
Target variable should have good correlation with independant varriables to get better results. Only highly correlated independant variable will be selected for prediction.

```python
# Check correlation for PerformanceRating
plt.rcParams["figure.figsize"] = 22,1
corrilation_depend = df.corr()
sns.heatmap(corrilation_depend.loc[["PerformanceRating"],:],annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

#### Insights
- EmpEnvironmentSatisfaction, EmpLastSalaryHikePercent, EmpWorkLifeBalance are Highly correlated with Performance Rating.
- ExperienceYearsAtThisCompany, ExperienceYearsInCurrentRole, YearsSinceLastPromotion, YearsWithCurrManager, EmpJobLevel are negatively correlated with Performance Rating.
